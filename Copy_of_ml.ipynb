```python
!pip install gradio -q
import gradio as gr
import pandas as pd
df = pd.read_csv("/content/enhanced_movie_dataset_1000.csv")
def recommend_movies(genre):
    genre = genre.lower()
    recs = df[df["genres"].str.lower().str.contains(genre)]

    if recs.empty:
        return "<h3> No movies found for this genre.</h3>"


    top = recs[["title", "imdb_rating", "popularity"]].sort_values(
        by="imdb_rating", ascending=False
    ).head(10)

    # HTML
    html = f"<h2 style='text-align:center; color:#ff5555;'>🔥 Top 10 {genre.title()} Movies</h2><br>"
    for _, row in top.iterrows():
        stars = "⭐" * int(round(row["imdb_rating"] / 2))
        html += f"""
        <div style='
            background-color:#202020;
            color:white;
            border-radius:15px;
            padding:15px;
            margin:10px 0;
            box-shadow:0 0 10px rgba(255,255,255,0.1);
        '>
            🎬 <b style='font-size:18px;'>{row['title']}</b><br>
            <span style='font-size:16px;'>{stars} ({row['imdb_rating']})</span><br>
            <span style='color:#ffaa00;'>🔥 Popularity: {row['popularity']}</span>
        </div>
        """
    return html

unique_genres = sorted(set("|".join(df["genres"].dropna()).split("|")))
with gr.Blocks(theme=gr.themes.Soft(primary_hue="red", secondary_hue="orange")) as app:
    gr.HTML("<h1 style='text-align:center; color:#ff4444;'>🎬 Movie Recommendation Dashboard</h1>")
    gr.Markdown(
        "Choose a genre to explore top-rated movies with IMDb ratings and popularity insights!"
    )

    with gr.Row():
        genre = gr.Dropdown(choices=unique_genres, label="🎞️ Choose Genre", value="Action", scale=2)
        btn = gr.Button("🔍 Get Recommendations", scale=1)

    output = gr.HTML(label="🎥 Recommended Movies")
    btn.click(recommend_movies, inputs=genre, outputs=output)

app.launch(share=True)
```
